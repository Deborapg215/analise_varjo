```python
import pandas as pd

vendas = pd.read_csv('Contoso - Vendas - 2017.csv', sep= ';')
clientes = pd.read_csv('Contoso - Clientes.csv', sep= ';', encoding='ISO-8859-1')
produtos = pd.read_csv('Contoso - Cadastro Produtos.csv', sep= ';', encoding='ISO-8859-1')

display(vendas)
display(clientes)
display(produtos)
```

```python
clientes = clientes[['ÿID Cliente', 'E-mail', 'Genero']]
clientes
```

```python
vendas = vendas.merge(produtos, on='ID Produto')
vendas
```

```python
display(clientes)
```

```python
clientes = clientes.rename(columns={'ÿID Cliente' : 'ID Cliente'})
clientes
```

```python
vendas = vendas.merge(clientes, on='ID Cliente')
vendas
```

```python
produtos = produtos.rename(columns={'ÿNome do Produto' : 'Nome do Produto'})
produtos
```

```python
vendas = vendas.rename(columns={'ÿNome do Produto' : 'Nome do Produto', 'E-mail' : 'E=mail do Cliente'})
vendas
```

```python
import pandas as pd 

display(vendas)
```

```python
# Qual cliente que comprou mais vezes?

frequencia_cliente = vendas['E=mail do Cliente']. value_counts()
display(frequencia_cliente)
frequencia_cliente[:10].plot(figsize=(15,5))
```

```python
# Qual produto que mais vendeu

produtos = produtos.groupby('Nome do Produto').sum()
display(produtos)

```

```python
produtos = produtos.sort_values('Quantidade Vendida', ascending=False)
display(produtos)

produtos [:5].plot(figsize=(15,5), kind='bar')
```

```python
# Produto que tem maior valor de vendas

maior_valor_vendas = produtos['Quantidade Vendida'].max()
menor_valor_vendas = produtos['Quantidade Vendida'].min()
produto_mais_vendido = produtos['Quantidade Vendida'].idxmax()
produto_menos_vendido = produtos['Quantidade Vendida'].idxmin()
print(maior_valor_vendas)
print(menor_valor_vendas)
print(produto_mais_vendido)
print(produto_menos_vendido)
```
